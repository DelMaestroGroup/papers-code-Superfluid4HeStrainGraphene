```python
from graphenetools import gt
```

```python
# Here we plot a funky box
strain = 0.00
iso_strain = 1.00
a = 1.42*iso_strain
ncellx = 8
ncelly = 8
lx = a*math.sqrt(3)*1 * ncellx
ly = a*1.5 * ncelly
box_dims = [-lx/2,lx/2,-ly/2,ly/2]

plt.style.reload_library()
fig, ax = gt.plot_graphene_lattice(strain,box_dims,carbon_carbon_distance=a)
dgutils.fonts.set_custom_font(font_path)

# Create a Rectangle patch
rect = patches.Rectangle((-lx/2*4/8, -ly/2*4/8), lx*4/8, ly*4/8, 
                         linestyle='dashed', linewidth=1, edgecolor='black', facecolor='none')

rect2 = patches.Rectangle((-lx/2*6/8, -ly/2*6/8), lx*6/8, ly*6/8, 
                         linestyle='dashed', linewidth=1, edgecolor='black', facecolor='none')

# Add the patch to the Axes
ax.add_patch(rect)
ax.add_patch(rect2)

fig.text(0.461, 0.689, 'N=16 Box', color='grey')
fig.text(0.46, 0.69, 'N=16 Box')
fig.text(0.461, 0.769, 'N=36 Box', color='grey')
fig.text(0.46, 0.77, 'N=36 Box')
fig.text(0.461, 0.849, 'N=64 Box', color='grey')
fig.text(0.46, 0.85, 'N=64 Box')

fig.set_dpi(300)

# fig.savefig('../figures/scale_of_bax.png', dpi=300)
```
