```python
# Here we plot a funky box
strain = 0.00
iso_strain = 1.00
a = 1.42*iso_strain
ncellx = 8
ncelly = 8
lx = a*np.sqrt(3)*1 * ncellx
ly = a*1.5 * ncelly
box_dims = [-lx/2,lx/2,-ly/2,ly/2]

plt.style.reload_library()
with plt.style.context(plot_style['aps']):
    #fig = plt.figure(constrained_layout=True ,figsize = (3.4646, 2.14122))
    fig, ax = gt.plot_graphene_lattice(strain,box_dims,carbon_carbon_distance=a)
    dgutils.fonts.set_custom_font(font_path)

    # Create a Rectangle patch
    rect = patches.Rectangle((-lx/2*4/8, -ly/2*4/8), lx*4/8, ly*4/8, 
                             linestyle='dashed', linewidth=1, edgecolor='black', facecolor='none')

    rect2 = patches.Rectangle((-lx/2*6/8, -ly/2*6/8), lx*6/8, ly*6/8, 
                             linestyle='dashed', linewidth=1, edgecolor='black', facecolor='none')

    # Add the patch to the Axes
    ax.add_patch(rect)
    ax.add_patch(rect2)

    fig.text(0.471, 0.660, r'$N_{\graphene}=16$', color='#525252', weight='bold')
    fig.text(0.4705, 0.661, r'$N_{\graphene}=16$', weight='bold')
    fig.text(0.471, 0.755, r'$N_{\graphene}=36$', color='#525252', weight='bold')
    fig.text(0.4705, 0.756, r'$N_{\graphene}=36$', weight='bold')
    fig.text(0.471, 0.849, r'$N_{\graphene}=64$', color='#525252', weight='bold')
    fig.text(0.4705, 0.8495, r'$N_{\graphene}=64$', weight='bold')

    fig.savefig('../figures/scale_of_box.pdf')
    fig.savefig('../figures/scale_of_box.svg')
# fig.set_dpi(300)

# fig.savefig('../figures/scale_of_box.png', dpi=300)
```
