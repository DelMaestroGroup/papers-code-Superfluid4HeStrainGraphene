```python
from graphenetools import gt
```

## Generating PIMC command line options for isotropic strain using Nathan's amazing library

```python
strain = 0.00
n=3
for i in range(0,15,2):
    iso_strain = 1+0.01*i
    a = 1.42*iso_strain
    print(i,":  ",end="")
    gt.roughly_square(n,strain,carbon_carbon_distance=a)
for i in range(15,24):
    iso_strain = 1+0.01*i
    a = 1.42*iso_strain
    print(i,":  ",end="")
    gt.roughly_square(n,strain,carbon_carbon_distance=a)
for i in range(24,31,2):
    iso_strain = 1+0.01*i
    a = 1.42*iso_strain
    print(i,":  ",end="")
    gt.roughly_square(n,strain,carbon_carbon_distance=a)
```

```python
strain = 0.00
iso_strain = 1.00
a = 1.42*iso_strain
n=3 # `(2n)^2` C1/3 adsorption sites
fig, ax = gt.roughly_square_plot(n,strain,carbon_carbon_distance=a)

gt.roughly_square(n,strain,carbon_carbon_distance=a) #print out the relevant PIMC parameters

# so we use this one!!!
# gt.c_one_third_commensurate_command(n,2*n,strain,carbon_carbon_distance=a)
fig.set_dpi(300)
```

## Use this if you want to specify a particular number of adsorption sites in the x- and y- directions

```python
# Here we plot a mostly rectangular plot commensurate with the C1/3 adsorption sites and print out the relevant PIMC parameters
strain = 0.00
iso_strain = 1.20
a = 1.42*iso_strain
m=2
n=4
fig, ax = gt.c_one_third_commensurate_command_plot(m,n,strain,carbon_carbon_distance=a)
gt.c_one_third_commensurate_command(m,n,strain,carbon_carbon_distance=a) #print out the relevant PIMC parameters
fig.set_dpi(300)
```
